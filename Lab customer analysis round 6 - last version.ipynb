```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df5 = df4.copy() # Remember 'total_claim_amount' was already dropped
numerical5 = df5.select_dtypes(include=np.number)
categorical5 = df5.select_dtypes(include=np.object)
```

```python
numerical5
```

```python
#Preprocess 'income' for removing zero values
```

```python
df6 = df5.copy()
log_lmbda = lambda x: np.log(x) if np.isfinite(np.log(x)) else np.NAN
df6['income'] = df6['income'].apply(log_lmbda)
df6['income'] = df6['income'].fillna(np.mean(df6['income']))
# The're way too many zero values which don't make sense to be replaced by the average. 
# Thus, converting zero values into NAN's make more sense than replacing them by the average
sns.distplot(df6['income'])
plt.show()
```

```python
numerical6 = df6.select_dtypes(include=np.number)
categorical6 = df6.select_dtypes(include=np.object)
numerical6
```

```python
numerical6_corr = numerical6.corr()
numerical6_corr
```

```python
sns.pairplot(numerical6) # remember 'total claim amount' was already dropped
plt.show()
```

```python
#numericals5_2 = X.select_dtypes(include = np.number)
#categorical5_2 = X.select_dtypes(include = np.object)
cat6_data = pd.get_dummies(categorical6, drop_first=True) # if you need to save the encoding info, this won't do
cat6_data
```

```python
#X_num = X.select_dtypes(include = np.number)
#X_cat = X.select_dtypes(include = np.object)
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(categorical6)
```

```python
encoded6 = encoder.transform(categorical6).toarray()
encoded6.shape 
```

```python
X = np.concatenate([numerical6, encoded6], axis=1)
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression() # Create squeleton
model.fit(X_train,y_train) # Coefficients for squeleton 
predictions  = model.predict(X_test) 
predictions
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_squared_error(y_test, predictions)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
X = sm.add_constant(X)

model = sm.OLS(y,X).fit() # to train the model

print(model.summary())
```

```python
lm = LinearRegression() # start the model object
model = lm.fit(X,y)
print(lm.coef_)
print(lm.intercept_)
```

```python
predictions = lm.predict(X) # predict some values

rmse = mean_squared_error(y, predictions, squared=False) # or mse with squared=True
# r2 gives you an overall estimate of how good the model is, between 0 (lowest) and 1 (highest)
# rmse, good to see if the errors are big, the values returnes are in the unit of the target
print("R2_score:", round(lm.score(X,y),3)) # or r2_score(Y, predictions)
print("RMSE:", rmse)
```

```python
predictions = lm.predict(X) # predict some values

mse = mean_squared_error(y, predictions, squared=True)
print("R2_score:", round(lm.score(X,y),3)) # or r2_score(Y, predictions)
print("MSE:", mse)
```
